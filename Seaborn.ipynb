```python
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri,style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri,style='event',markers=True)
plt.show()
```

Distrubutiion plot

like whise in the matplotlib has histogram attribute but in this seaborn we use it as distrubution plot(distplot)

```python
import seaborn as sns
```

```python
diamond=pd.read_csv('diamond.csv')
```

```python
diamond.head()
```

```python
sns.distplot(diamond['price'])
plt.show()
```

```python
sns.distplot(diamond['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamond['price'],color='brown',kde=True)
plt.show()
```

```python
sns.distplot(diamond['price'],color='brown',kde=False,bins=10)
plt.show()
```

```python
sns.distplot(diamond['price'],color='brown',vertical=True)
plt.show()
```

Joint plot

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,color='olive')
plt.show()
```

Using new attribution called kind for the regression line for both histogram and scatter plot 

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,color='olive',kind='reg')
plt.show()
```

```python
churn=pd.read_csv('churn.csv')
```

```python
churn.head()
```

```python
sns.boxplot(y='tenure',x='Churn',data=churn)
plt.show()
```

```python
sns.boxplot(y='MonthlyCharges',x='InternetService',data=churn)
plt.show()
```

```python
sns.boxplot(y='tenure',x='Contract',data=churn,palette='Set1')
plt.show()
```

```python
sns.boxplot(y='tenure',x='Contract',data=churn,linewidth=3)
plt.show()
```

```python
sns.boxplot(y='tenure',x='Contract',data=churn,order=['Two year','Month-to-month','One year'])
plt.show()
```

```python
sns.boxplot(y='tenure',x='Contract',data=churn,order=['Two year','Month-to-month','One year'],hue='PaymentMethod',palette='Set1')
plt.show()
```
