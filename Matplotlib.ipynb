```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,100,10)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.title("First line plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Bar ploting

```python
students={'sam':30,'chai':40,'nag':50}

names=list(students.keys())
marks=list(students.values())
```

```python
plt.bar(names,marks)
plt.show()
```

```python
plt.bar(names,marks)
plt.title("Marks of students")
plt.xlabel("students")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,marks,color='green')
plt.title("Marks of students")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.grid(True)
plt.show()
```

Scatter ploting

```python
x= (10,20,30,40,50,60,70,80)
a=(5,6,3,8,2,1,2,7)
```

```python
plt.scatter(x,a)
plt.show()
```

```python
plt.scatter(x,a,marker='*',color='orange',s=300)
plt.show()
```

```python
x= (10,20,30,40,50,60,70,80)
a=(5,6,3,8,2,1,2,7)
b=(3,1,7,9,4,6,3,2)
```

```python
plt.scatter(x,a,marker="*",c="b",s=200)
plt.scatter(x,b,marker=".",c="orange",s=300)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="b",s=200)
plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="orange",s=300)
plt.show()
```

```python
fruits=('banana','orange','apple','pomagranate','grapes')
quantity=(91,15,60,40,20)
```

before ploting we have to give the numerical object that is quantity and then catogorical object

```python
plt.pie(quantity,labels=fruits)
plt.show()
```

```python
fruits=('banana','orange','apple','pomagranate','grapes')
quantity=(91,15,60,40,20)
plt.pie(quantity,labels=fruits,autopct="%0.2f%%",colors=['yellow','orange','red','pink','green'])
plt.show()
```

in the above pie chat we used the autopct=%o.2f%% we can also use o.1 , this is an atribute which gives us the numerical values on the chat.

```python
plt.pie(quantity,labels=fruits,autopct="%0.2f%%",colors=['yellow','orange','red','pink','green'],radius=2)
plt.pie([5],radius=1,colors='w')
plt.show()
```
